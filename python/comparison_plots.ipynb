```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("results/comparison.csv", index_col=0)
print(df)
```

```python
cities = df.index
plt.figure(figsize=(8, 4), dpi=150)
plt.xticks(cities)
plt.plot(cities, df['seq'], marker='o', alpha=0.8)
plt.plot(cities, df['omp'], marker='o', alpha=0.8)
plt.plot(cities, df['cuda_v1'], marker='o', alpha=0.8)
plt.plot(cities, df['cuda_v2'], marker='o', alpha=0.8)
plt.legend(['sekwencyjnie', 'openmp', 'cuda v1', 'cuda v2'], loc="upper left")
plt.grid(axis='x')
plt.xlabel("Liczba miast")
plt.ylabel('czas [s]')
```

```python
cities = df.index
plt.figure(figsize=(8, 4), dpi=150)
plt.xticks(cities)
plt.yscale("log") 
plt.plot(cities, df['seq'], marker='o', alpha=0.8)
plt.plot(cities, df['omp'], marker='o', alpha=0.8)
plt.plot(cities, df['cuda_v1'], marker='o', alpha=0.8)
plt.plot(cities, df['cuda_v2'], marker='o', alpha=0.8)
plt.legend(['sekwencyjnie', 'openmp', 'cuda v1', 'cuda v2'], loc="upper left")
plt.grid(axis='x')
plt.xlabel("Liczba miast")
plt.ylabel('czas [s]')
```

```python
cities = df.index
plt.figure(figsize=(8, 4), dpi=150)
plt.xticks(cities) 
plt.plot(cities, df['seq'] / df['omp'] , marker='o', alpha=0.8)
plt.plot(cities, df['seq'] / df['cuda_v1'] , marker='o', alpha=0.8)
plt.plot(cities, df['seq'] / df['cuda_v2'] , marker='o', alpha=0.8)
plt.legend(['openmp', 'cuda v1', 'cuda v2'], loc="upper left")
plt.grid(axis='x')
plt.xlabel("Liczba miast")
plt.ylabel('stosunek czasu')
```
