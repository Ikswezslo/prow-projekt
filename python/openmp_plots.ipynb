```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("results/omp_result.csv", index_col=0)
#print(df)
```

```python
plt.figure(figsize=(8, 4), dpi=150)
plt.xticks(df.index)
plt.bar(df.index,df['n300'])
plt.bar(df.index,df['n250'])
plt.bar(df.index,df['n200'])
plt.legend(['n = 300', 'n = 250', 'n = 200'], loc="upper right")
plt.xlabel("Liczba wƒÖtk√≥w (0 oznacza program sekwencyjny)")
plt.ylabel('Czas dzia≈Çania [s]')
```

```python
plt.figure(figsize=(8, 4), dpi=150)
X = [int(x[1:]) for x in df.columns]
plt.xticks(X )
plt.bar(X, df.loc[1] - df.loc[0], 20)
plt.xlabel("Liczba miast")
plt.ylabel('R√≥znica czas√≥w dzia≈Çania [s]')
```

```python
threads = df.index[1:]
plt.figure(figsize=(8, 4), dpi=150)
plt.xticks(threads)
speedup = df.n300[0] / df.n300[1:]
plt.plot(threads, speedup, marker='o', alpha=0.8)
speedup = df.n200[0] / df.n200[1:]
plt.plot(threads, speedup, marker='o', alpha=0.8)
speedup = df.n100[0] / df.n100[1:]
plt.plot(threads, speedup, marker='o', alpha=0.8)
plt.legend(['n = 300', 'n = 200', 'n = 100'], loc="lower right")
plt.grid(axis='x')
plt.xlabel("Liczba wƒÖtk√≥w")
plt.ylabel('Przyspieszenie')
```

```python
threads = df.index[1:]
plt.figure(figsize=(8, 4), dpi=150)
plt.xticks(threads)
speedup = df.n300[0] / df.n300[1:]
efficiency = speedup / threads
plt.plot(threads, efficiency, marker='o', alpha=0.8)
speedup = df.n200[0] / df.n200[1:]
efficiency = speedup / threads
plt.plot(threads, efficiency, marker='o', alpha=0.8)
speedup = df.n100[0] / df.n100[1:]
efficiency = speedup / threads
plt.plot(threads, efficiency, marker='o', alpha=0.8)
plt.legend(['n = 300', 'n = 200', 'n = 100'], loc="upper right")
plt.grid(axis='x')
plt.xlabel("Liczba wƒÖtk√≥w")
plt.ylabel('Wydajno≈õƒá')
```

```python
threads = df.index[1:]
plt.figure(figsize=(8, 4), dpi=150)
plt.xticks(threads)
speedup = df.n300[0] / df.n300[1:]
karp_flatt = (1 / speedup - 1 / threads) / (1 - 1 / threads)
plt.plot(threads[1:], karp_flatt[1:], marker='o', alpha=0.8)
speedup = df.n200[0] / df.n200[1:]
karp_flatt = (1 / speedup - 1 / threads) / (1 - 1 / threads)
plt.plot(threads[1:], karp_flatt[1:], marker='o', alpha=0.8)
speedup = df.n100[0] / df.n100[1:]
karp_flatt = (1 / speedup - 1 / threads) / (1 - 1 / threads)
plt.plot(threads[1:], karp_flatt[1:], marker='o', alpha=0.8)
plt.legend(['n = 300', 'n = 200', 'n = 100'], loc="upper right")
plt.grid(axis='x')
plt.xlabel("Liczba wƒÖtk√≥w")
```
